```python
def read_datas(dir):
    return pd.read_csv(dir)
```

```python
def draw_plot(title,x,y,xlabel,ylabel):
    plt.plot(x,y)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()
```

```python
input_addr = "/kaggle/input/global-population/Global_annual_population.csv"
data_csv = read_datas(input_addr)
```

```python
years = data_csv.iloc[:,0]
population = data_csv.iloc[:,1]
```

```python
draw_plot("Global Population",years,population,"Years","Population (billion)")
```
